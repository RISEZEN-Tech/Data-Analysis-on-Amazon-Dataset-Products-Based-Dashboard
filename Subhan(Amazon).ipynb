```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
```

```python
df = pd.read_csv(r"C:\Users\Mian Yamin\Downloads\sana_amazon.csv")
df.head()
```

```python
df = df.drop(columns = ["product_name", "user_id", "user_name", "review_id", "review_content", "img_link", "about_product", "discount_percentage" ,"product_link"])
df.head()
```

```python
df["discounted_price"] = df["discounted_price"].str.replace("₹", "")
df["actual_price"] = df["actual_price"].str.replace("₹", "")
df["rating_count"] = df["rating_count"].str.replace(",", "")
df["actual_price"] = df["actual_price"].str.replace(",", "")
df["discounted_price"] = df["discounted_price"].str.replace(",", "")
df.head()
```

```python
le = LabelEncoder()
df["product_id"] = le.fit_transform(df["product_id"]).astype(int)
df["category"] = le.fit_transform(df["category"]).astype(int)
df["category"] = le.fit_transform(df["category"]).astype(int)
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df["discounted_price"] = pd.to_numeric(df["discounted_price"], errors = "coerce")
fill_disc = df["discounted_price"].median()
df["discounted_price"] = df["discounted_price"].fillna(fill_disc)
df["discounted_price"] = df["discounted_price"].astype(int)

df["actual_price"] = pd.to_numeric(df["actual_price"], errors = "coerce")
fill_act_price = df["actual_price"].median()
df["actual_price"] = df["actual_price"].fillna(fill_act_price)
df["actual_price"] = df["actual_price"].astype(int)

df["rating"] = pd.to_numeric(df["rating"], errors = "coerce")
fill_rating = df["rating"].median()
df["rating"] = df["rating"].fillna(fill_rating)

df["rating_count"] = pd.to_numeric(df["rating_count"], errors = "coerce")
fill_rat_count = df["rating_count"].median()
df["rating_count"] = df["rating_count"].fillna(fill_rat_count)
df["rating_count"] = df["rating_count"].astype(int)
df.head()
```

```python
df["Price After Discount"] = df["actual_price"] - df["discounted_price"]
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df[["product_id", "category",	"discounted_price", "actual_price", "rating", "rating_count"]]
y = df[["Price After Discount"]]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
model.predict([[346	,10, 399, 1099, 4.2, 24269]])
```

```python
from sklearn.metrics import mean_squared_error , r2_score, accuracy_score
mse = mean_squared_error(y_test, y_predict)
score = r2_score(y_test, y_predict)
```

```python
print(f"Mean Squared Value: {mse:.2f}")
print(f"R2 Score is: {score:.2f}")
```

```python
mape = np.mean(np.abs((y_predict - y_test) / y_test.replace(0, np.finfo(float).eps))) * 100
accuracy = 100 - mape
print(f"Approximate Accuracy: {accuracy:.2f}%")
```
